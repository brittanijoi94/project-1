```python
#Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts
```

```python
#Csv read

gdp_income_df=pd.read_csv("EXCEL_gdp_income.csv")
gdp_income_df
```

```python
#Replacing $ and , 

gdp_income_df["GDP (Dollars)"]=gdp_income_df["GDP (Dollars)"].str.replace('$', '')
gdp_income_df["GDP (Dollars)"]=gdp_income_df["GDP (Dollars)"].str.replace(',', '')
gdp_income_df["GDP (Dollars)"]=gdp_income_df["GDP (Dollars)"].astype(float)
gdp_income_df


```

```python
gdp_income_df["Population (No. of People)"]=gdp_income_df["Population (No. of People)"].str.replace('$', '')
gdp_income_df["Population (No. of People)"]=gdp_income_df["Population (No. of People)"].str.replace(',', '')
gdp_income_df["Population (No. of People)"]=gdp_income_df["Population (No. of People)"].astype(int)
gdp_income_df
```

```python
gdp_income_df["Disposable Personal Income (Millions of Dollars)"]=gdp_income_df["Disposable Personal Income (Millions of Dollars)"].str.replace('$', '')
gdp_income_df["Disposable Personal Income (Millions of Dollars)"]=gdp_income_df["Disposable Personal Income (Millions of Dollars)"].str.replace(',', '')
gdp_income_df["Disposable Personal Income (Millions of Dollars)"]=gdp_income_df["Disposable Personal Income (Millions of Dollars)"].astype(float)
gdp_income_df
```

```python
gdp_income_df["Personal Income (Millions of Dollars)"]=gdp_income_df["Personal Income (Millions of Dollars)"].str.replace('$', '')
gdp_income_df["Personal Income (Millions of Dollars)"]=gdp_income_df["Personal Income (Millions of Dollars)"].str.replace(',', '')
gdp_income_df["Personal Income (Millions of Dollars)"]=gdp_income_df["Personal Income (Millions of Dollars)"].astype(float)
gdp_income_df
```

```python
gdp_income_df["Per Capita Disposable Personal Income (Dollars)"]=gdp_income_df["Per Capita Disposable Personal Income (Dollars)"].str.replace('$', '')
gdp_income_df["Per Capita Disposable Personal Income (Dollars)"]=gdp_income_df["Per Capita Disposable Personal Income (Dollars)"].str.replace(',', '')
gdp_income_df["Per Capita Disposable Personal Income (Dollars)"]=gdp_income_df["Per Capita Disposable Personal Income (Dollars)"].astype(int)
gdp_income_df
```

```python
gdp_income_df["Per Capita Personal Income (Dollars)"]=gdp_income_df["Per Capita Personal Income (Dollars)"].str.replace('$', '')
gdp_income_df["Per Capita Personal Income (Dollars)"]=gdp_income_df["Per Capita Personal Income (Dollars)"].str.replace(',', '')
gdp_income_df["Per Capita Personal Income (Dollars)"]=gdp_income_df["Per Capita Personal Income (Dollars)"].astype(int)
gdp_income_df
```

```python
#Median for all columns 
median=pd.DataFrame({"GDP Mean":[gdp_income_df["GDP (Dollars)"].median()],
"Pop Median":[gdp_income_df["Population (No. of People)"].median()],
"DPI Median":[gdp_income_df["Disposable Personal Income (Millions of Dollars)"].median()],
"Personal Median":[gdp_income_df["Personal Income (Millions of Dollars)"].median()],
"PC Dis Median":[gdp_income_df["Per Capita Disposable Personal Income (Dollars)"].median()],
"PC Per In Median":[gdp_income_df["Per Capita Personal Income (Dollars)"].median()]})
median
```

```python
#using gdp_income_df to create visuals

plt.scatter(gdp_income_df["GDP (Dollars)"],gdp_income_df["Disposable Personal Income (Millions of Dollars)"],edgecolors="black",facecolors="darkgray")
plt.title("GDP vs. Disposable Personal Income (Millions of Dollars)")
plt.xlabel("GDP")
plt.ylabel("Disposable Personal Income (Millions of Dollars)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.ticklabel_format(style = 'plain')
plt.savefig("Figure1.png")
plt.show()
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('United States GDP, 1969-2019')
ax1.set_ylabel('GDP (Dollars)')
ax1.boxplot(gdp_income_df['GDP (Dollars)'])
plt.savefig("Figure2.png")
plt.show()
```

```python
gdp_line=gdp_income_df.plot('Year', 'GDP (Dollars)', color="black", label="GDP")
plt.title("Annual United States GDP, 1969-2019")
plt.xlabel("Year")
plt.ylabel("GDP (Dollars)")
plt.ticklabel_format(style = 'plain')
plt.savefig("Figure3.png")
plt.show()
```

```python
PCD_line=gdp_income_df.plot('Year', 'Per Capita Disposable Personal Income (Dollars)', color="black", label="Per Capita Disposable Personal Income (Dollars)")
plt.title("Annual Per Capita Disposable Personal Income (Dollars), 1969-2019")
plt.xlabel("Year")
plt.ylabel("Per Capita Disposable Personal Income (Dollars)")
plt.ticklabel_format(style = 'plain')
plt.savefig("Figure4.png")
plt.show()
```
