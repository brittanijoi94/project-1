```python
import pandas as pd
```

```python
gdp1_xls = "GDP_1969_1997.xls"
gdp2_xls = "GDP_1997_2019.xls"
```

```python
gdp1_df = pd.read_excel(gdp1_xls, index_col=None, header=None)
gdp2_df = pd.read_excel(gdp2_xls, index_col=None, header=None)
```

```python
gdp1 = gdp1_df.dropna().reset_index(drop=True)
gdp1 = gdp1.drop(columns=[0])
gdp1 = gdp1.set_index([1])
gdp1 = gdp1.transpose()
gdp1 = gdp1.rename(columns={"GeoName": "Years", "United States": "GDP"})
gdp1["Years"] = gdp1["Years"].astype(int)
gdp1
```

```python
gdp2 = gdp2_df.dropna().reset_index(drop=True)
gdp2 = gdp2.drop(columns=[0])
gdp2 = gdp2.set_index([1])
gdp2 = gdp2.transpose()
gdp2 = gdp2.rename(columns={"GeoName": "Years", "United States *": "GDP"})
gdp2["Years"] = gdp2["Years"].astype(int)
gdp2
```

```python
gdp_complete = pd.concat([gdp1, gdp2], ignore_index=True)
gdp_complete = gdp_complete.reset_index(drop=True)
gdp_complete = gdp_complete.drop(gdp_complete.index[0:6])


gdp_complete.to_excel("GDP_1969_2020.xlsx", index=False)
gdp_complete
```

```python
gdp_complete.to_csv("GDP_1969_2020.csv", index=False)
```
